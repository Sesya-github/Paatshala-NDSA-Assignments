# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Reading the dataset 

```python
data=pd.read_excel('iris.xls')
data
```

# Meaningful visualizations on dataset


```python
plt.figure(figsize=(8,8))
sns.boxplot(x='Classification',y='SL',data=data)
plt.title('Sepal length based on classifications',size=16)
plt.ylabel('Sepal Length')
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='PL',y='PW',data=data,hue='Classification')
plt.title('Petal width v/s Petal length based on classifications',size=16)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
plt.figure(figsize=(8,8))
sns.violinplot(x='Classification',y='SW',data=data)
plt.title('Sepal width based on classifications',size=16)
plt.ylabel('Sepal Width')
```
